This lesson tackles eligibility traces as a new algorithmic mechanism to achieve more in our online updates. This mechanism allows us to efficiently implement RL algorithms that perform n-step bootstrapped learning without waiting for or accessing the previous n steps. Instead, we maintain a set of parameters updated regularly online each time an agent visits a state. Then this vector representing recently visited states will be sued to update the weights instead of the simple state vector. Eligibility traces are a powerful trick often used in classical RL.

# Prediction with Eligibility Traces

## TD(位)
TD(位) uses eligibility traces which allow us to implement the 位-return algorithm more efficiently in a straightforward and online manner! Refer to section 12.2. This is an online algorithm. Surprisingly, we can obtain the average of an offline set of all n-step TD updates. The equivalence is only guaranteed for the offline case. Nevertheless, the online case works just fine. The next section will tackle the true online, which addresses this shortcoming. 
The accumulating eligibility trace takes the form. 

$z_{t} = \gamma\lambda z_{t-1} + \hat{v}(S_t, w_t)$

while for the update, we replace the gradient with the eligibility vector. Note that this type of eligibility trace has the same gradient requirement regarding storage and processing time.

$\delta_{t} = R_{t+1} + \gamma \hat{v}(S_{t+1},w_t) - \hat{v}(S_t,w_t)$

$w_{t+1} = w_{t} + \alpha\delta_{t}z_{t}$

The eligibility traces need to be initialised at the start of each episode, and we achieve that by using our step0 function.

```python
class TD位(vMRP):
    def __init__(self, 位=.5, **kw):
        super().__init__(**kw)
        self.位 = 位
    
    def step0(self):
        self.z = self.w*0

    # -----------------------------  online learning ----------------------    
    def online(self, s, rn,sn, done, *args): 
        伪, 纬, 位 = self.伪, self.纬, self.位
        self.z = 位*纬*self.z + self.V(s)
        self.w += 伪*(rn + (1-done)*纬*self.V(sn) - self.V(s))*self.z
        
```

```python
TD位walk = TD位(env=vrandwalk(), 位=.9, 伪=.03, episodes=100, v0=.5, **demoV()).interact(label='TD learning')
```

```python
TDwalk = TD(env=vrandwalk(), 伪=.03, episodes=100,  v0=.5, **demoV()).interact(label='TD learning')
```

As we can see TD(位) can converge faster than TD as it compresses several TD updates in each of its updates. This leads to the jittery behaviour that usually makes us reduce the learning rate but we can achieve more updates in less episodes in general. The other downside is that we have to tune the hyper parameter 位 which can take more effort. 

## True Online TD(位) and Dutch Traces
We implement the True online TD(位) algorithm in this section. This algorithm guarantees an equivalence between the online and the offline cases, and it was hailed as an important step towards implementing a multi-step prediction algorithm that is both efficient and sound. in particular, it guarantees an equivalence between the offline 位-return algorithm (which takes the average of all possible n-step updates) and the true online TD(位)! The justification or proof is quite involved. Please refer to sections 12.5 and 12.6 of the book. you can also have a look at the original true TD [paper](https://proceedings.mlr.press/v32/seijen14.html) and subsequent [paper](https://www.jmlr.org/papers/volume17/15-599/15-599.pdf) the latter is packed with useful information.

The drawback is that it can only be applied on a linear approximation and not on any general one as TD(位), but it is a step in the right direction until researchers find a way to do it also for the nonlinear function approximator as well (such as neural networks). 

We need to handle an eligibility trace variable (Dutch trace), the size of which is identical to the weights. The update rule is given in equation 12.11, and its justification/deduction can be read from section 12.5. which shows the formula for Dutch traces.

Dutch traces are available when we use linear function approximation. They perform better than the classical accumulated eligibility traces. They take the form of 

$z_{t} = \gamma\lambda z_{t-1} + (1-\alpha\gamma\lambda z_{t-1}^\top x_{t})x_{t}$

The update takes the form:

$\delta_{t} = R_{t+1} + \gamma \hat{v}(S_{t+1},w_t) - \hat{v}(S_t,w_t)$

$w_{t+1} = w_{t} + \alpha\delta_{t}z_{t} + \alpha( w_{t}^\top x_{t} - w_{t-1}^\top x_{t})(z_{t} - x_{t})$

When we implement the true online TD(位) algorithm, we save computation by saving the result of $vo=w_{t-1}^\top x_{t}$ instead of storing $w_{t-1}$.

```python
class trueTD位(vMRP):
    def __init__(self, 位=.5, **kw):
        super().__init__(**kw)
        self.位 = 位

    def step0(self):
        self.z = self.w*0
        self.vo = 0
    # -----------------------------  online learning ----------------------    
    def online(self, s, rn,sn, done, *args): 
        伪, 纬, 位 = self.伪, self.纬, self.位
        
        self.v = self.V(s)
        self.vn= self.V(sn)*(1-done)
        未 = rn + 纬*self.vn - self.v
        self.z = 位*纬*self.z + (1-伪*位*纬*self.z.dot(s))*s
        
        self.w += 伪*(未 + self.v - self.vo )*self.z - 伪*(self.v - self.vo)*s
        self.vo = self.vn
        
```

```python
trueTD位walk = trueTD位(env=vrandwalk(), 伪=.03, 位=.8, episodes=100, v0=.5, **demoV()).interact(label='TD learning')
```

```python
trueTD位walk = trueTD位(env=vrandwalk(), 伪=.03, 位=.4, episodes=100, v0=.5, **demoV()).interact(label='TD learning')
```

```python
TDwalk = TD(env=vrandwalk(), episodes=100, 伪=.03,  v0=.5, **demoV()).interact(label='TD learning')
```

As we can see true online TD(位) can converge faster than TD. Again, similar to TD(位) this leads to the jittery behaviour that usually entails reducing the learning rate, but we can achieve more updates in less episodes in general. The other downside is that we have to tune the hyper parameter 位 which can take more effort. ### Model selection for trueTD位: Tuning 位 and 伪

We will test our true online TD on the random walk with function approximation. Due to the dot product, some of the learning rate 伪 might be too much for the algorithm when using high 位 values, and they will cause overflow. This is not a problem as we are just testing the limits of the algorithm, but to avoid doing so, we added a conditional statement to limit the values of 伪 for high 位 values.

```python
def TD位_MC_Walk_伪compare(algorithm=TD位, label='TD(位)', runs=10):
    
    steps0 = list(np.arange(.001,.01,.001))
    steps1 = list(np.arange(.011,.2,.02))
    steps2 = list(np.arange(.25,1.,.05))

    伪s = np.round(steps0 +steps1 + steps2, 2)
    #伪s = np.arange(0,1.05,.1) # quick testing
    
    plt.xlim(-.02, 1)
    plt.ylim(.24, .56)
    plt.title('%s RMS error Average over 19 states and first 10 episodes'%label)
    for 位 in [0, .1, .4, .8, .9, .95, .975, .99, 1]:
        end=34 if 位<.975 else (-3 if 位<.99 else -10)
        compare = Compare(algorithm=algorithm(env=vrandwalk_(), v0=0, 位=位, episodes=10), 
                                  runs=runs, 
                                  hyper={'伪':伪s[:end]}, 
                                  plotE=True).compare(label='位=%.3f'%位)

    if algorithm==trueTD位:
        compare = Compare(algorithm=MC(env=vrandwalk_(), v0=0, episodes=10), 
                                  runs=runs, 
                                  hyper={'伪':伪s}, 
                                  plotE=True).compare(label='MC  TD(位=1)', frmt='-.')
    
```

# Control with Eligibility Traces
We move next to the control realm and write a Sarsa(位) with an approximate value function. The idea is similar to Sarsa(0) that we have written in the previous lesson (we called is just Sarsa there). We need, however, this time to take care of an extra variable that holds the eligibility traces, which has a size identical to the weights, so each action has its own eligibility trace. 

The eligibility traces need to be initialised at the start of each episode, and we achieve that by using our step0 function. The update is per equation 12.5 in the book, but we must apply it for the eligibility trace corresponding to the current action.

Note that the book assumes that we concatenate all the eligibility traces by encoding a large state representation that assigns 0 to all components that do not correspond with the current action. We felt that this was vague and made the implementation unclear.

## Sarsa(位)

```python
class Sarsa位(vMDP):
    def __init__(self, 位=.5, **kw):
        super().__init__(**kw)
        self.位 = 位
        self.step = self.step_an # for Sarsa we want to decide the next action in time step t
    
    def step0(self):
        self.Z = self.W*0
    # ---------------------------------------- online learning ----------------------------------------
    def online(self, s, rn,sn, done, a,an):
        self.Z[a] = self.位*self.纬*self.Z[a] + self.Q(s)
        self.W[a] += self.伪*(rn + (1-done)*self.纬*self.Q(sn,an)- self.Q(s,a))*self.Z[a]
```

## True Online Sarsa(位)

We implement the True online Sarsa(位) algorithm in this section. This algorithm guarantees an equivalence between the online and the offline cases, and it is hailed as an important step towards implementing a sound multi-step control algorithm that is both efficient and sound. The main drawback that we have is that it can only be applied to a linear approximation, not to any general one as Sarsa(位), but it is a step in the right direction until researchers find a way to do it also for the nonlinear functions as well (such as neural networks). We need to handle an eligibility trace variable, the size of which is identical to the weights, so each action has its own eligibility trace. The update rule is given in equation 12.11, and its justification/deduction can be read from section 12.5.
Note that in our implementation. 

- qo is an old action-value estimation for the current state
- qn is the action-value estimation for the next    state
- q  is the action-value estimation for the current state


```python
class trueSarsa位(vMDP):
    def __init__(self, 位=.5, **kw):
        super().__init__(**kw)
        self.位 = 位
        self.step = self.step_an # for Sarsa we want to decide the next action in time step t
    
    def step0(self):
        self.Z = self.W*0
        self.qo = 0
    # ---------------------------------------- online learning ----------------------------------------
    def online(self, s, rn,sn, done, a,an):
        
        伪, 纬, 位 = self.伪, self.纬, self.位
        
        self.q = self.Q(s,a)
        self.qn= self.Q(sn,an)*(1-done)
        未 = rn + 纬*self.qn - self.q
        self.Z[a] = 位*纬*self.Z[a] + (1-伪*位*纬*self.Z[a].dot(s))*s
        
        self.W[a] += 伪*(未 + self.q - self.qo )*self.Z[a] - 伪*(self.q - self.qo)*s
        self.qo = self.qn
```
